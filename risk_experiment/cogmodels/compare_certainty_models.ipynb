```python
import arviz as az
```

```python
certainty_lapse_3t2 = az.from_netcdf('/data/ds-risk/derivatives/cogmodels/ses-3t2_model-certainty_lapse_trace.netcdf')
certainty_lapse_7t2 = az.from_netcdf('/data/ds-risk/derivatives/cogmodels/ses-7t2_model-certainty_lapse_trace.netcdf')

certainty_full_3t2 = az.from_netcdf('/data/ds-risk/derivatives/cogmodels/ses-3t2_model-certainty_full_trace.netcdf')
certainty_full_7t2 = az.from_netcdf('/data/ds-risk/derivatives/cogmodels/ses-7t2_model-certainty_full_trace.netcdf')

certainty_evidence_3t2 = az.from_netcdf('/data/ds-risk/derivatives/cogmodels/ses-3t2_model-certainty_evidence_trace.netcdf')
certainty_evidence_7t2 = az.from_netcdf('/data/ds-risk/derivatives/cogmodels/ses-7t2_model-certainty_evidence_trace.netcdf')
```

```python
comparison_3t2 = az.compare({'certainty (lapse) - 3t2':certainty_lapse_3t2,
            'certainty (evidence) - 3t2':certainty_evidence_3t2,
            'certainty (full) - 3t2':certainty_full_3t2,
             })
```

```python
az.plot_compare(comparison_3t2)
```
