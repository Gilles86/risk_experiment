```python
import numpy as np
import pandas as pd
# Sample randomly in log space between log(5) and log(28):

n = 10000
safe_payoffs = np.exp(np.random.uniform(low=np.log(5), high=np.log(28), size=n))
fractions = np.random.uniform(low=1, high=4, size=n)
risk_payoffs = safe_payoffs * fractions

# Create a dataframe with the payoffs
payoffs = pd.DataFrame({'safe_payoff': safe_payoffs, 'risk_payoff': risk_payoffs})

payoffs
```

```python
print(payoffs.describe())

expected_payoff = .5 * payoffs['safe_payoff'] + .55 * .5 * payoffs['risk_payoff']

expected_payoff.describe()
```

```python
def get_symbolic_trials(n=128):

    n_safe = np.random.rand(n) * 23
    n_safe += 5

    frac = np.random.rand(n)*3 + 1

    n_risky = n_safe * frac

    return pd.DataFrame({'n_safe':np.round(n_safe,2), 'n_risky':np.round(n_risky, 2)})
    
```

```python
sns.scatterplot(data=get_symbolic_trials(), x='n_safe', y='n_risky')
```
